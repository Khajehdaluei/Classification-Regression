```python
from fbprophet import Prophet
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../__data/AirPassengers.csv')
data.head()
```

```python
data['Month'] = pd.DatetimeIndex(data['Month'])
data.set_index('Month').plot(figsize=(20, 12))
```

```python
p = Prophet()

data = data.rename(columns={'Month': 'ds', 'AirPassengers': 'y'})   # columes name for fit model
p.fit(data)
```

```python
future = p.make_future_dataframe(periods=36, freq='MS')

forecast = p.predict(future)

p.plot(forecast, uncertainty=True)
```
